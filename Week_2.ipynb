```python
# Mean for numeric
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

# Correct way to fill missing values without using inplace=True
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Department'] = df['Department'].fillna(df['Department'].mode()[0])

print(df)
```

```python
#Forward Fill
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

df_ffill = df.ffill()

print(df_ffill)
```

```python
#backward Fill
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

df_bfill = df.bfill()

print(df_bfill)
```

```python
#Deletion
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

df_drop_rows = df.dropna()
print("After dropping rows:\n", df_drop_rows)
print("-" * 30)

df_drop_cols = df.dropna(axis=1)
print("After dropping columns:\n", df_drop_cols)
```

```python
#Exact Match Removal
import pandas as pd

df = pd.DataFrame({
    'ID': [1, 2, 2, 3, 4, 4],
    'Name': ['Alice', 'Bob', 'Bob', 'Charlie', 'David', 'David'],
    'Age': [25, 30, 30, 35, 40, 40]
})

print("Original Data:\n", df)
print("-" * 30)

df_exact = df.drop_duplicates()
print("After Exact Match Removal:\n", df_exact)
```

```python
#Subset-Based Removal
import pandas as pd

df = pd.DataFrame({
    'ID': [1, 2, 2, 3, 4, 4],
    'Name': ['Alice', 'Bob', 'Bob', 'Charlie', 'David', 'David'],
    'Age': [25, 30, 30, 35, 40, 40]
})

print("Original Data:\n", df)
print("-" * 30)

df_subset_id = df.drop_duplicates(subset=['ID'])
print("After Subset-Based Removal (ID):\n", df_subset_id)
print("-" * 30)

df_subset_name = df.drop_duplicates(subset=['Name'])
print("After Subset-Based Removal (Name):\n", df_subset_name)
```

```python
#Date Standardization
import pandas as pd

# Sample dataset with inconsistent date formats
df = pd.DataFrame({
    'Date': ['2025-01-05', '05/01/2025', 'Jan 5, 2025', '2025.01.05']
})

# Convert all dates to standard ISO format safely
df['Date'] = pd.to_datetime(df['Date'], format='mixed', errors='coerce').dt.strftime('%Y-%m-%d')

print(df)
```

```python
#Case Normalization
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'BOB', 'charlie', 'DAVID']
})

df['Name_lower'] = df['Name'].str.lower()

df['Name_upper'] = df['Name'].str.upper()

print(df)
```
