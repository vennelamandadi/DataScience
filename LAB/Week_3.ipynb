```python
# Euclidean Distance
import numpy as np
from scipy.spatial import distance

pointA = np.array([1, 2, 3])
pointB = np.array([4, 6, 8])

euclidean_dist = distance.euclidean(pointA, pointB)
print("Euclidean Distance:", euclidean_dist)

similarity_euclidean = 1 / (1 + euclidean_dist)
print("Euclidean Similarity:", similarity_euclidean)
```

```python
#Manhattan Distance
import numpy as np
from scipy.spatial import distance

# Define the sample points (matching your Euclidean example)
pointA = np.array([1, 2, 3])
pointB = np.array([4, 6, 8])

# Calculate Manhattan Distance
# Formula: $\sum |A_i - B_i|$
manhattan_dist = distance.cityblock(pointA, pointB)
print("Manhattan Distance:", manhattan_dist)

# Calculate Similarity (inverse of distance)
# Maps the distance to a range of (0, 1]
similarity_manhattan = 1 / (1 + manhattan_dist)
print("Manhattan Similarity:", similarity_manhattan)
```

```python
# Minkowski Distance with p=3
import numpy as np
from scipy.spatial import distance
minkowski_dist_p3 = distance.minkowski(pointA, pointB, p=3)
print("Minkowski Distance (p=3):", minkowski_dist_p3)

# Similarity (inverse of distance)
similarity_minkowski = 1 / (1 + minkowski_dist_p3)
print("Minkowski Similarity (p=3):", similarity_minkowski)
```

```python
#Pearson Correlation
import pandas as pd
import seaborn as sns
from scipy.stats import pearsonr

# Load the dataset directly from seaborn's online repository
tips = sns.load_dataset("tips")

# Calculate Pearson correlation between 'total_bill' and 'tip'
corr, p_value = pearsonr(tips['total_bill'], tips['tip'])

print(f"Pearson correlation coefficient: {corr:.4f}")
print(f"P-value: {p_value}")
```

```python
#Spearman Rank Correlation
import pandas as pd
from scipy.stats import spearmanr

# Example dataset
df = pd.DataFrame({
    'X': [10, 20, 30, 40, 50],
    'Y': [12, 18, 33, 47, 55]
})

# Spearman correlation coefficient and p-value
corr_value, p_value = spearmanr(df['X'], df['Y'])
print(f"Spearman Correlation Coefficient: {corr_value}")
print(f"P-value: {p_value}")
```

```python
#Hamming Distance
import numpy as np
from scipy.spatial import distance

# Define two numeric arrays
arr1 = np.array([1, 0, 1, 1, 0])
arr2 = np.array([1, 1, 0, 1, 0])

# scipy returns the percentage of positions that differ (2 out of 5 = 0.4)
scipy_dist = distance.hamming(arr1, arr2)
print("Scipy Hamming Distance (proportion):", scipy_dist)

# To get the absolute count of differing elements:
absolute_dist = scipy_dist * len(arr1)
print("Absolute Hamming Distance:", int(absolute_dist))
```

```python
# Cosine Similarity
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

s1 = "data science is fun"
s2 = "science makes data useful"

vectorizer = CountVectorizer().fit([s1, s2])
vectors = vectorizer.transform([s1, s2])

cos_sim = cosine_similarity(vectors[0], vectors[1])[0][0]
print("Cosine Similarity:", cos_sim)
```

```python
# Sequence-Based Measures
def lcs_length(X, Y):
    m, n = len(X), len(Y)

    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(m):
        for j in range(n):

            if X[i] == Y[j]:
                dp[i + 1][j + 1] = dp[i][j] + 1

            else:
                dp[i + 1][j + 1] = max(dp[i][j + 1], dp[i + 1][j])

    return dp[m][n]

seq1 = "ABCDEF"
seq2 = "AEBDF"

length = lcs_length(seq1, seq2)

print("Longest Common Subsequence Length:", length)
```
